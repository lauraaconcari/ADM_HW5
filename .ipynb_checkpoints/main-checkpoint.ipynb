```python
def fill_nodes_citation(data, G):
    for _, row in data.iterrows():
        paper_id = row['id']
        references_str = row['references']

        # Convert the references string to a list of integers
        references = [int(ref) for ref in str(references_str).split(';') if ref.isdigit()]

        # Add edges to the graph if the reference exists in the dataset
        for ref in references:
            if ref in data['id'].values:
                G.add_edge(paper_id, ref)
```

#### **Functionality 1 - Graph's features**

Input:
* The graph
* The name of the graph

Output:
* The number of the nodes in the graph
* The number of the edges in the graph
* The graph density
* The graph degree distribution
* The average degree of the graph
* The graph hubs (hubs are nodes having degrees more extensive than the 95th percentile of the degree distribution)
* Whether the graph is dense or sparse

In order to returns the required graph features we implement the function graph_feature() that uses the networkx library

```python
#definition of the function that implements functionality 1

def graph_features(g, g_name):
    """This function returns some features of the input graph."""
    
    # Number of nodes
    g_nodes = g.number_of_nodes()
    
    # Number of edges
    g_edges = g.number_of_edges()
    
    # Graph density
    g_density = nx.density(g) if g.number_of_edges() > 0 else 0
    
    # Graph degree distribution
    g_degree_distr = [i for i, freq in enumerate(nx.degree_histogram(g)) for _ in range(freq)]
    
    # Graph average degree
    degree_view = pd.DataFrame(g.degree, columns=['Node', 'Degree'])
    degree_array = np.array(degree_view['Degree'])
    g_average_degree = np.mean(degree_array)
    
    # Graph hubs
    percentile_95 = np.percentile(degree_array, 95)
    g_hubs = degree_view.loc[degree_view['Degree'] > percentile_95, 'Node'].values
    
    # Dense or sparse
    g_dense_or_sparse = 'Dense' if g_density > 0.5 else 'Sparse'
    
    # Construction of a dataframe
    columns = ['Number of nodes', 'Number of edges', 'Graph density', 'Graph degree distribution', 'Graph average degree', 'Graph hubs', 'Graph type of density']
    g_features_list = [(int(g_nodes), int(g_edges), g_density, g_degree_distr, g_average_degree, g_hubs, g_dense_or_sparse)]
    g_features_df = pd.DataFrame(g_features_list, columns=columns, index=[g_name])
    
    # Returns all the features required and also a dataframe that contains them
    return g_name, g_nodes, g_edges, g_density, g_degree_distr, g_average_degree, g_hubs, g_dense_or_sparse, g_features_df
```

#### **Functionality 2 - Nodes' contribution**

Input:

* The graph
* A node of the graph (paper/author)
* The name of the graph

Output:

* The centrality of the node, calculated based on the following centrality measurements:
    * Betweeness
    * PageRank
    * ClosenessCentrality
    * DegreeCentrality

In this cell we define the function used to compute some centrality measurement given a graph and a node as input

```python
def nodes_contribution(g, n, g_name):
    """Function to understand the importance of a certain node in the graph, based on some centrality measures
    """
    # Check if the node exists in the graph
    if n not in g:
        return f"The node {n} is not present in the graph {g_name}."
    
    # Calculate centrality measures
    betweeness_centr = nx.betweenness_centrality(g)[n]
    pagerank_centr = nx.pagerank(g)[n]
    closeness_centr = nx.closeness_centrality(g, u=n)
    degree_centr = nx.degree_centrality(g)[n]
    
    # Return all the measurements required and also a dataframe that contains them
    return g_name, n, betweeness_centr, pagerank_centr, closeness_centr, degree_centr
```

```python
#definition of the function that returns the subgraph of the top N nodes based on degree
def top_N_auth(g,N):
    """This function returns a subgraph of the input graph containing the top N nodes with most degree"""
    #create a dictionary with nodes as keys and degree as values
    dict_degree = dict(g.degree())
    
    #create a pandas Series from the previous dictionary sorted in descending order containing the top N nodes with most degree
    S_degree = pd.Series(dict_degree).sort_values(ascending=False).head(N)
    
    #create a list containing the top N nodes
    top_nodes = list(S_degree.index)
    
    #create a subgraph of the input graph containing only the top N nodes
    g_subgraph = g.subgraph(top_nodes).copy()
    
    return g_subgraph
```

The function shortest_ordered_walk compute the shortest ordere path between the given authors and returns the number of steps and the crossed papers

```python
#definition of the function that implements functionality 3
def shortest_ordered_walk(g, authors_a, start_end_points, N):
    if g != collaboration_graph:
        return 'No walk calculated, you should insert the Collaboration graph'
    
    # Create the subgraph of top N nodes using the function "top_N_auth"
    g_sub = top_N_auth(g, N)

    # Check if the subgraph is not connected
    if not nx.is_connected(g_sub):
        return 'No walk calculated, the subgraph is not connected'

    # Divide the start and the end points
    start_point, end_point = start_end_points

    # Create a list containing all the points in order
    all_points = [start_point] + authors_a + [end_point]
    cross_papers = []
    shortest_walk = []

    for i in range(len(all_points) - 1):
        path, pap = BFS_shortest_path(g_sub, all_points[i], all_points[i + 1])
        
        if type(path) == str:
            return 'No path'
            
        shortest_walk += path[1:]
        cross_papers += pap

    return shortest_walk, cross_papers
```

```python
def shortest_ordered_walk(graph, nodes_in_order, start_node, end_node):
    def dfs(current_path):
        current_node = current_path[-1]

        if current_node == end_node and len(current_path) == len(nodes_in_order) + 2:
            return current_path

        for neighbor in graph.get(current_node, {}):
            if neighbor not in current_path:
                if current_path[-1] == nodes_in_order[0]:
                    if neighbor == nodes_in_order[1]:
                        new_path = dfs(current_path + [neighbor])
                        if new_path:
                            return new_path
                elif neighbor == nodes_in_order[len(current_path) - 1]:
                    new_path = dfs(current_path + [neighbor])
                    if new_path:
                        return new_path

    path = dfs([start_node])
    return path
```

#### **Functionality 4 - Disconnecting Graphs**

Input:
* The graph data
* authorA: a paper to which will relate sub-graph G_a
* authorB: a paper to which will relate sub-graph G_b
* $N$: denoting the top $N$ authors that their data should be considered

Output:
* The minimum number of edges (by considering their weights) required to disconnect the original graph in two disconnected subgraphs: G_a and G_b.

```python
#importing
from networkx.algorithms.flow import edmonds_karp
```

The function below calculate the neighbours of the given list of nodes

```python
def neighbors_list(graph, list_of_nodes):
    neighbors = set()
    for node in list_of_nodes:
        neighbors.update(graph.neighbors(node))
    return neighbors
```

This function implement the BFS algorithm

```python
def bfs_custom(graph, starting_node):
    queue = [starting_node]
    visited = set(queue)

    while queue:
        node = queue.pop(0)
        for neighbor in graph.neighbors(node):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return visited
```

This function disconnecting_graphs() implements the functionality 4 using the Edmonds-Karp algorithm and the previous defined BFS

```python
def disconnecting_graphs(graph, auth_a, auth_b, N):
    # Apply the top_N_auth function to get the subgraph.
    subgraph = top_N_auth(graph, N)
    
    # Apply the Edmonds-Karp algorithm to find the residual network.
    residual = edmonds_karp(subgraph, auth_a, auth_b, 'weight')
    
    # Use BFS to find all nodes reachable from auth_a in the residual graph.
    reachable_nodes = bfs_custom(residual, auth_a)
    
    # Determine the cut sets based on the BFS exploration.
    set1 = set(reachable_nodes)
    set2 = set(graph.nodes) - set1
    separating_edges = list(nx.edge_boundary(graph, set1, set2))
    
    # Return the size of the cut (minimum number of edges), the actual edges, and the modified graph.
    return len(separating_edges), separating_edges, graph
```

#### **Functionality 5 - Extracting Communities**

Input:
* The graph data
* $N$: denoting the top $N$ papers that their data should be considered
* Paper_1: denoting the name of one of the papers
* Paper_2: denoting the name of one of the papers

Output:
* The minimum number of edges that should be removed to form communities
* A list of communities, each containing a list of papers that belong to them.
* Whether the Paper_1 and Paper_2 belongs to the same community.


Function to create a subgraph by considering only the top N nodes (or authors) based on their degree centrality in the original graph.

```python
def top_N_nodes(G, N):
    degrees = dict(G.degree())
    N_authors = sorted(degrees.items(), key=lambda x: x[1], reverse=True)[:N]
    N_authors_id = [el[0] for el in N_authors]
    subgraph = G.subgraph(N_authors_id)
    return subgraph
```

This function identifies the edge with the highest edge betweenness centrality in a given graph

```python
def edge_to_remove(graph):
    G_dict = nx.edge_betweenness_centrality(graph)
    edge = ()

    # Extract the edge with the highest edge betweenness centrality score
    for key, value in sorted(G_dict.items(), key=lambda item: item[1], reverse=True):
        edge = key
        break

    return (len(edge), edge)
```

This function applies the Girvan-Newman algorithm to detect communities in a graph by iteratively removing edges with the highest edge betweenness centrality

```python
def girvan_newman(graph):
    # Find the number of connected components
    sg = nx.connected_components(graph)
    sg_count = nx.number_connected_components(graph)

    # Continue removing edges until the graph is divided into multiple connected components
    while sg_count == 1:
        edge = edge_to_remove(graph)[1]
        graph.remove_edge(edge[0], edge[1])
        sg = nx.connected_components(graph)
        sg_count = nx.number_connected_components(graph)

    return sg
```

This function checks whether two given papers belong to the same community in a list of communities

```python
def same_community(Paper_1, Paper_2, communities):
    for community in communities:
        if Paper_1 in community and Paper_2 in community:
            return True
    return False
```

This function extracts communities from a graph and checks if two specified papers belong to the same community

```python
def extracting_communities(graph, Paper_1, Paper_2, N):
    if N:
        graph = top_N_nodes(graph, N)
    
    # compute number of edges    
    number_of_edges = edge_to_remove(graph)[0]

    c = girvan_newman(graph.copy())

    # find the nodes forming the communities
    communities = []

    for i in c:
        communities.append(list(i))
    
    same_c = same_community(Paper_1, Paper_2, communities)
    
    return (number_of_edges, communities, same_c)
```

#### **Visualization 2 - Visualize node's contribution**

Centrality is a concept in network analysis that measures the importance of nodes within a graph. There are various centrality measurements, and the phrase "node's centrality value based on the four centrality measurements" suggests the use of four specific metrics to assess the significance of nodes in a network.

The four common centrality measurements are:

* **Degree Centrality**: It quantifies how many connections a node has. Nodes with higher degree centrality are more connected within the network.

* **Closeness Centrality**: It evaluates how close a node is to all other nodes in the network. Nodes with higher closeness centrality are more accessible in terms of reaching other nodes quickly.

* **Betweenness Centrality**: This measures the extent to which a node lies on the shortest paths between other nodes. Nodes with higher betweenness centrality act as bridges or critical connectors in the network.

* **Eigenvector Centrality**: It assesses a node's importance by considering the importance of its neighbors. Nodes connected to other well-connected nodes have higher eigenvector centrality.

In the context of the visualization, the "node's centrality value based on the four centrality measurements" implies calculating and presenting numerical values for each node in the graph using these four centrality metrics. This information helps to identify and rank nodes based on different aspects of their importance or influence within the network.

### Part A

#### 1. Implement an algorithm to solve the described mentioned problem.


```python
# Input N, M and S. After this is done it let's you add to the list_skills all the skills required
N, M, S = map(int, input().split())
list_skills = list(map(str, input().split()))


# Initialize athlete_skills_list in which we will save all the id and skills of the atheltes
athlete_skills_list = list()


# Appends to the athlete_skills_list all the dictionaries computed
for i in range(N):
  athlete_id = int(input())
  skills = {}

  # for loop to record skill and level of each athlete
  for i in range(S):
      skill, level = input().split()
      level = int(level)
      skills[skill] = level

  # Now we record id and skill of each athlete
  athlete_skills_list.append({'id': athlete_id, 'skills': skills})


def calculate_max_score(required_skills, athlete_skills_list):
  # Create a list of tuples (skill, score, athlete_id) for all athletes
  all_skills = []
  for athlete in athlete_skills_list:
    athlete_id = athlete['id']
    for skill, score in athlete['skills'].items():
      all_skills.append((skill, score, athlete_id))

# Now we sort the skills by score in descending order to keep only the highest ones
all_skills.sort(key=lambda x: x[1], reverse=True)

# Keep track of used athletes
used_athletes = set()
skill_max_scores = []

# Iterate over required skills and find the best available athlete for each
for skill in required_skills:
  for athlete_skill in all_skills:
  if athlete_skill[0] == skill and athlete_skill[2] not in used_athletes:
    skill_max_scores.append(athlete_skill[1])
    used_athletes.add(athlete_skill[2])
    break
  else:
    # If no athlete has the required skill we don't add anything
    skill_max_scores.append(0)

return sum(skill_max_scores)


print(calculate_max_score(list_skills, athlete_skills_list))
```

### Part B

#### 1. Prove or disprove that the problem is NP-complete.


This problem has two components:

*   Covering all skills in the set T
*   Minimizing the sum of edge weights to work well together

To show it is NP-complete we need to show that:
-  The problem is in NP --> Solution to be verified in polynomial time:

  For each skill in T we need to check if there is at least one member in V' who has that skill. This is done by iterating over the set T and X and this can be done in O(T * V').

  Secondly, we can use a minimum spanning tree to check if V' is minimized. To do this, we first check if the tree connects all vertices in V' without any cycle. Then, check if the total weight of the tree is less than or equal of any other form of G[V']. This can be done using Kruskal algorithm which as a time-complexity of O(V' log V).

  Both of these steps are in polynomial-time form so we can assert that the problem is NP.

-  The problem is NP-Hard: We can show this by taking a known NP-complete problem and transforming it into our problem in such a way that if we can solve our problem, we can also solve the original NP-complete problem. (https://ics.uci.edu/~eppstein/161/960312.html). One of these examples of NP-complete problems is the Set Cover Problem (https://math.mit.edu/~goemans/18434S06/setcover-tamara.pdf), where we have U elements and we need to find the smallest subset of S such taht every element in U is included. In our case, we have a set of skills T, same as to the universe U in the Set Cover problem, and a group of individuals X, each with a subset of these skills, similar to each set in S in the Set Cover problem. Our goal is to find the subset of these individuals such that all skills in T are covered. To show that our problem is NP-hard, we can map each element in the universe U of the Set Cover problem to a skill in T, and each set in S to an individual in X. We therefore assign a weight to the edges between individuals to represent the effort to work together, simplifying the reduction to focus on skill coverage. If we can find a subset of individuals X′ that covers all skills with the minimized sum of edge weights, we effectively solve the Set Cover, demonstrating that out problem is at least as hard as the Set Cover problem, confirming that our problem is NP-hard.

Combined with the fact that a solution can be verified in polynomial time, this establishes the NP-completeness of the problem.
